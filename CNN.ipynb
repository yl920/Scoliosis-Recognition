```python
import torch
import torch.nn as nn
```

```python
class MyCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.cnn = nn.Sequential(
            nn.Conv2d(1,32,3), #input, output, size
            nn.MaxPool2d(2,2), #size, stride
            nn.Conv2d(32,64,3),
            nn.MaxPool2d(2,2),  #Convolution+Pooling
            nn.Linear(6,4), #Fully connected+ Activation
            nn.ReLU(),
            nn.Linear(4,2),
            nn.ReLU(),
            nn.Linear(2,1)
        )
    def forward(self,x):
        y = self.cnn(x)
        return y
```

```python
x = torch.randn(1,32,32)
model = MyCNN()
y = model(x)
print(y.shape,y)
```
